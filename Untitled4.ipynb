```python
import os
```

```python
path="C:\\Program Files (x86)\Python36-32"
df=pd.read_csv(path+'/country_iso_code/STEAM_country.csv')
country_path=path+'/country_clean/raw/'
cont=[]
contfull=[]
for con in os.listdir(country_path):
    if con !='.DS_Store':
        cont.append(con[:2])
        name=df["official_name_en"].loc[df["ISO3166-1-Alpha-2"].str.contains(con[:2].upper())==True]
        contfull.extend(name)
```

```python
populs=[]
temp=pd.read_csv(path+"/country.csv")
for i in range(len(cont)):
    populs.extend(temp["Population"].loc[temp["Country Name"]==contfull[i]])
```

```python
ppps=[]
temp=pd.read_csv(path+"/ppp.csv")
for i in range(len(cont)):
    ppps.extend(temp["PPP"].loc[temp["Country Name"]==contfull[i]].astype(float))
```

```python
popul=dict(zip(cont,populs))
ppp=dict(zip(cont,ppps))
iso3=[]
sales=[]
sales_cs=[]
prices_cs=[]
for country in cont:
    iso3.extend(df['ISO3166-1-Alpha-3'].loc[df["ISO3166-1-Alpha-2"].str.lower()==country])
    sales_year_cs=0
    frame=pd.read_csv(path+"/country_clean/country_appid/"+country+".csv")
    sales_year_cs=frame["sales_q"].loc[frame["name"]=="Counter-Strike: Global Offensive"]
    prices_cs.extend(frame["local_price_usd"].loc[frame["name"]=="Counter-Strike: Global Offensive"]/ppp[country])
    sales_cs.extend(sales_year_cs/popul[country])

```

```python
data=pd.DataFrame({"country":cont,"sales_cs":sales_cs,"prices_cs":prices_cs})
data.head()
```

```python
p=figure(height=200)
p.circle(x=data.prices_cs,y=data.sales_cs)
show(p)
```

```python
from bokeh.models import HoverTool, ColumnDataSource, LinearInterpolator, LinearColorMapper, ColorBar
from bokeh.palettes import viridis
Spectral=viridis(68)
color_mapper=LinearColorMapper(palette=Spectral,low=data.sales_cs.min(),high=data.sales_cs.max())
size_mapper=LinearInterpolator(x=[min(populs),max(populs)],y=[5,50])
source=ColumnDataSource(dict(x=data.prices_cs,y=sales_cs,country=data.country,population=populs))
hover=HoverTool(tooltips="@country",show_arrow=False)
p=figure(tools=[hover],height=200,title="Counter Strike")
p.circle(x="x",y="y",size={"field":"population","transform":size_mapper},source=source,alpha=0.6,color={"field":"y","transform":color_mapper})
p.add_layout(ColorBar(color_mapper=color_mapper,location=(0,0),orientation="vertical"),"right")
show(p)
output_file("plot1.html")
```
